# Building Dataset

```python
import os, shutil

data_dir = '/media/shan/Data/places365/places3_standard'
source_dir = '/media/shan/Data/places365/places365standard_easyformat/places365_standard'

interested_places= ['office',
                    'office_cubicles',
                    'supermarket',
                    'shopping_mall-indoor',
                    'lobby',
                    'elevator_lobby',
                    'museum-indoor',
                    'general_store-indoor',
                    'subway_station-platform',
                    'airport_terminal',
                    'art_gallery',
                    'hotel_room',
                    'hospital_room',
                    'waiting_room']
```

```python
### Train dataset

num_places = len(interested_places)
image_per_place = int(5000 / num_places)
train_path_list = []

for place_name in interested_places:
    counter = 0
    for filename in os.listdir(os.path.join(source_dir, 'train', place_name)):
        if counter >= image_per_place:
            continue
        if filename.endswith(".jpg"): 
            train_path_list.append(os.path.join(source_dir, 'train', place_name, filename))
            counter += 1
            # print(os.path.join(source_dir, 'train', place_name, filename))
print("## Train dataset")
print("Number of image per place: {0}".format(image_per_place))
print("Number of places interested: {0}".format(num_places))
print("Number of path collected: {0}".format(len(train_path_list)))

print("Copying train dataset ...")
filename_counter = 0
for path in train_path_list:
    target_path = os.path.join(data_dir, 'train', 'other', '{0}.jpg'.format(filename_counter))
    shutil.copyfile(path, target_path)
    filename_counter += 1
print("Done copying {0} images".format(filename_counter))
```

```python
### Validation dataset

num_places = len(interested_places)
image_per_place = int(100 / num_places)
val_path_list = []

for place_name in interested_places:
    counter = 0
    for filename in os.listdir(os.path.join(source_dir, 'val', place_name)):
        if counter >= image_per_place:
            continue
        if filename.endswith(".jpg"): 
            val_path_list.append(os.path.join(source_dir, 'val', place_name, filename))
            counter += 1
            # print(os.path.join(source_dir, 'val', place_name, filename))

print("## Val dataset")
print("Number of image per place: {0}".format(image_per_place))
print("Number of places interested: {0}".format(num_places))
print("Number of path collected: {0}".format(len(val_path_list)))

print("Copying val dataset ...")
filename_counter = 0
for path in val_path_list:
    target_path = os.path.join(data_dir, 'val', 'other', '{0}.jpg'.format(filename_counter))
    shutil.copyfile(path, target_path)
    filename_counter += 1
print("Done copying {0} images".format(filename_counter))
```

## Data visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline

print("Sample images from val images from 'other' class")

img, vaxis = plt.subplots(5,4, figsize=(15, 20))
vaxis = vaxis.ravel()
idx = 0

for filename in os.listdir(os.path.join(data_dir, 'train', 'other')):
    if idx >= 20:
        break
    image = plt.imread(os.path.join(data_dir, 'train', 'other', filename))
    vaxis[idx].imshow(image)
    vaxis[idx].axis('off')
    idx += 1
```

## Split dataset

```python
import os, shutil, random

src_dir = '/media/shan/Data/places365/places3_standard/valtest'
val_dir = '/media/shan/Data/places365/places3_standard/val'
test_dir = '/media/shan/Data/places365/places3_standard/test'
for classname in os.listdir(src_dir):
    for filename in os.listdir(os.path.join(src_dir, classname)):
        if filename.endswith(".jpg"): 
            image_path = os.path.join(src_dir, classname, filename)
            target_path = ""
            
            if random.randint(0, 100) < 50:
                target_path = os.path.join(val_dir, classname, filename)
            else:
                target_path = os.path.join(test_dir, classname, filename)
                
            shutil.copyfile(image_path, target_path)
            print(target_path)
```
